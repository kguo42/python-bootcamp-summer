# Debugging: Error Messages and Understanding Them

Believe it or not this is the easiest form of debugging that can be done. The reason for this is that the computer gives you an error message that you can then interpret and back track to pinpoint where the error occured, what caused it and figure out a way to fix it. In this section we will cover some of the most common error messages that we have encountered in our research and give some advice on how to fix the issue. 

Some of the most common error we have seen is:

    SyntaxError: Happens when Python syntax is incorrect, this can be missing a : in the for loop declaration 
                 for example, or forgetting the close the paranthesis in print [ie: print('hello' ]
    IndentationError: Incorrect indentation in one part of the code
    SyntaxError: EOL while scanning string literal (You forgot to close the quotations in your statement)
    NameError: Typically happens when you misspell a variable and it will let you know that the variable 
               name is not defined
    IndexError: list index out of range (The index you provided is bigger than the length of the list or array)
    FileNotFoundError: This can happen when you are reading in a file and it cannot find it, this can be due to 
                       a misspelled file or the file path is incorrect. For this make sure the file path is 
                       pointing to the correct directory and that the entirety of the the path is correct.

# Specutils

Specutils is an amazing python package and is primarily most useful when you are working with an spectroscopic data. This has great documentations and highly recommend if you are suing spectral data to give this a read through and use the functions in specutils to aid in your research.

Recommend following the link here to learn more about Specutils: https://specutils.readthedocs.io/en/stable/index.html 

```python
from astropy.io import fits
from astropy import units as u
import numpy as np
from matplotlib import pyplot as plt
from astropy.visualization import quantity_support
quantity_support()  # for getting units on the axes below  
```

```python
filename = 'https://data.sdss.org/sas/dr16/sdss/spectro/redux/26/spectra/1323/spec-1323-52797-0012.fits'
# The spectrum is in the second HDU of this file.
with fits.open(filename) as f:  
    specdata = f[1].data  
```

```python
#Specutils main advantage is the Spectrum1D object that makes a spectrum that specutils can work with
from specutils import Spectrum1D

#Note the introduction of units in u.AA and u.Unit('ergs/s/cm^2/A')
lamb = 10**specdata['loglam'] * u.AA 
flux = specdata['flux'] * 10**-17 * u.Unit('erg cm-2 s-1 AA-1') 
spec = Spectrum1D(spectral_axis=lamb, flux=flux) 
```

```python
f, ax = plt.subplots(figsize = (12, 6))  
ax.step(spec.spectral_axis, spec.flux) 
```

```python
import warnings
from specutils.fitting import fit_generic_continuum

#this is a function to fit the continuum of the spectrum using the fit_generic_continuum function
with warnings.catch_warnings():  # Ignore warnings
    warnings.simplefilter('ignore')
    #normalizing the spectrum by the continuum
    cont_norm_spec = spec / fit_generic_continuum(spec)(spec.spectral_axis) 
```

```python
f, ax = plt.subplots()  
ax.step(cont_norm_spec.wavelength, cont_norm_spec.flux)  
ax.set_xlim(654 * u.nm, 660 * u.nm)  
```

# Photutils

Photutils is another amazing astronomy package and deals with measuring photometry from a 2D image. It has loads of sub-packages that can detect sources, perform aperture photometry and much more. I highly recommend checking out the documentation and exploring it to find something that you may end up using in your research. 

Link to Photutils: https://photutils.readthedocs.io/en/stable/#getting-started

```python
from photutils.datasets import make_100gaussians_image

data = make_100gaussians_image()
```

```python
from photutils.background import Background2D, MedianBackground
bkg_estimator = MedianBackground()
bkg = Background2D(data, (50, 50), filter_size=(3, 3),
                   bkg_estimator=bkg_estimator)
data -= bkg.background  # subtract the background
```

```python
threshold = 1.5 * bkg.background_rms
```

```python
from astropy.convolution import convolve
from photutils.segmentation import make_2dgaussian_kernel
kernel = make_2dgaussian_kernel(3.0, size=5)  # FWHM = 3.0
convolved_data = convolve(data, kernel)
```

```python
from photutils.segmentation import detect_sources
segment_map = detect_sources(convolved_data, threshold, npixels=10) #detect sources 
print(segment_map)
#This is an example to create a starting-point segmentation map, 
#in practice, please work with the term to explore best ways to do back ground substraction, estimating parameters, etc.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import SqrtStretch
from astropy.visualization.mpl_normalize import ImageNormalize
norm = ImageNormalize(stretch=SqrtStretch())

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12.5))
ax1.imshow(data, origin='lower', cmap='Greys_r', norm=norm)
ax1.set_title('Background-subtracted Data')
ax2.imshow(data, origin='lower', cmap='Greys_r', norm=norm)
ax2.contour(segment_map, origin='lower', levels =segment_map.labels, cmap=segment_map.cmap,
           interpolation='nearest', lw = 1)
ax2.set_title('Segmentation Image')
```

# Exception Handeling

Exception handeling is a cool way of making sure your code still runs even if an error pops up. Let's say that you are trying to perform a calculation and encounter a division by zero error, you could remove the zero from your array but another thing you can do is perform exception handeling to handle this. The way you do Exception Handeling is by using the Try-Except syntax. 


    try:
        Code to try here
        
    except SOME_ERROR_HERE:
        Do this if SOME_ERROR_HERE pops up
        
    except SOME_OTHER_ERROR_HERE:
        Do this if that error pops up
    

```python
values = [1, 3, 4, 6, 100000]

for vals in values:
    
    try:
        RA = FIELD_DF.RA.values
        Get_RA = RA[vals]
        ra = FIELD_DF.loc[vals, 'RA']
        
        print(f'RA Aquired for index: {vals}')
        print(f'RA is: {ra:.4f}')
    except KeyError:
        print('KeyError Triggered')
        print(f'Index {vals} not in DF')
    except IndexError:
        print('IndexError Triggered')
        
```
